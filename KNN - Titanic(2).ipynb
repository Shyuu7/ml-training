```python
import pandas as pd

#importei o dataset localmente pois não achei uma forma de utilizá-lo diretamente
df = pd.read_csv('train.csv', index_col="PassengerId")

#leitura inicial dos dados
df
```

```python
#checando as entradas sem valores
nan_check = df.isna().sum()
nan_check
```

```python
#muitos dados faltantes em cabines (687 de 891 entradas), melhor descartarmos esta coluna
df = df.drop('Cabin', axis=1)
df
```

```python
#vamos completar as idades faltantes com a mediana das idades dos passageiros (melhor do que média pois haverá apenas valores inteiros)
df['Age'] = df['Age'].fillna(df['Age'].median())
df
```

```python
#O nome do passageiro e o número do ticket também não parecem ser relevantes ao estudo, então vamos retirá-las do df
df = df.drop('Name', axis=1)
df = df.drop('Ticket', axis=1)
df
```

```python
#agora, vamos transformar as variáveis categóricas em numéricas. precisamos fazer isto pras colunas 'sex' e 'embarked'
#gêneros: female = 0 e male = 1 // portos: C = 0, Q = 1, S = 2

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Sex'] = le.fit_transform(df['Sex'])
df['Embarked'] = le.fit_transform(df['Embarked'])
df
```

```python
#finalmente podemos separar nossas features e nosso target

X = df.drop('Survived', axis=1)
y = df['Survived']

#checando nossas features
X
```

```python
#embora os dados estejam praticamente na mesma escala, vamos escaloná-los para ter certeza de que não haverá desbalanço entre as features
#optei pelo standardScaler por não achar nada de especial sobre os dados, então nenhum cuidado adicional parece ser necessário

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
#finalmente podemos separar os conjuntos de treino e teste para montarmos nosso modelo
#como solicitado, temos 80% dos dados para treino e 20% para teste

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
#agora vamos aplicar o algoritmo KNN. mantive o k padrão sugerido (5) para a primeira tentativa

from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
y_pred
```

```python
#por fim, vamos testar a acurácia do modelo através da comparação entre o y predito e o y real

from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print(f"Acurácia de {(accuracy * 100):.2f}%")
```

```python
#será que conseguimos a melhor acurácia possível? vamos testar outros valores para o k e checar as diferentes acurácias

for k in range(1, 50, 2):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Acurácia de {(accuracy * 100):.2f}% para k = {k}")
```

Em conclusão: 5 não parece ser o melhor valor para k neste modelo. Se usarmos k = 15 ou k = 29, conseguimos uma acurácia de 81,56%. Dentre os dois, eu optaria pelo k = 15 para reduzir o risco de underfitting.

```python
#vamos mudar nosso k para 15

knn = KNeighborsClassifier(n_neighbors=15)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
```

```python
#criando uma representação visual de quais predições o modelo acertou
import matplotlib.pyplot as plt
import numpy as np

# criando o gráfico
plt.figure(figsize=(10, 6))
plt.scatter(range(len(y_test)), y_test, label="Valores reais", marker="o")
plt.scatter(range(len(y_test)), y_pred, label="Valores preditos", marker="x")

# adicionando nomes aos eixos
plt.xlabel("Index")
plt.ylabel("Sobreviveu? (0 ou 1)")
plt.title("Comparação entre sobrevivência real vs. predita")

# mostrando a legenda
plt.legend()

# mostrando o gráfico
plt.show()
```

```python
#o gráfico não ficou muito claro por haver muitas marcações muito próximas umas das outras. vamos fazer uma matriz de confusão para visualizar os dados com números absolutos
from sklearn.metrics import confusion_matrix
import seaborn as sns

cm = confusion_matrix(y_test, y_pred)

sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predito')
plt.ylabel('Real')
plt.show()
```
