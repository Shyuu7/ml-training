a. Explique brevemente como funciona o algoritmo de árvore de decisão para classificação.

b. Em que aspectos o Random Forest melhora o desempenho em relação a uma única árvore?

c. Quais são os possíveis impactos do aumento excessivo no número de árvores em uma Random Forest?


R:
a. Uma árvore de decisão para classificação é construído da seguinte forma: com base nas features apresentadas, testes lógicos são criados (representados como nós). Cada possível resposta a esse nó representa um ramo da árvore, que liga o nó a um outro nó (caso mais testes lógicos são necessários) ou a uma folha, que representa a classe final e o fim daquele ramo da árvore. Uma amostra é inserida no nó inicial (chamado de nó raiz) e passa pelos testes lógicos sequencialmente até chegar ao nó folha com a sua classe correspondente. Esse modelo visa tentar separar ao máximo os dados da amostragem em cada nó, algo que pode ser medido com o índice de "pureza" - Gini ou Entropia.


---



b. O Random Forest é um modelo mais robusto e mais consistente que uma única árvore de decisão pois trabalha com conceitos de maioria (nos casos de classificação) e média (nos casos de regressão) do aglomerado de árvores treinado. Ao treinar múltiplos modelos menos complexos (porém diferentes entre si por conta da aleatoriedade dos dados e das features), estamos tentando balancear o fato de que os modelos são mais propensos a erros, porém raramente modelos diferentes cometem os mesmos erros, então no geral estes erros tendem a se anular. Também torna o modelo menos propenso a overfitting, pois é um modelo construído em cima de vários outros, e não dependente de um único modelo. Desta forma, possui uma variância menor.

---

c. Ao aumentar excessivamente o número de componentes de um Random Forest estaremos aumentando exponencialmente o custo computacional do modelo e lentificando o processo de treinamento, porém também estaremos reduzindo a variância e o risco de overfitting (até um certo ponto, pois o modelo estabiliza e adicionar mais árvores não traz mais benefícios visíveis).

## Questão 2 - SVM e Kernels:



a. Explique o papel do kernel em um modelo SVM. O que muda ao usar um kernel linear, RBF ou polinomial?

b. Como os parâmetros C e gamma afetam a complexidade do modelo e a margem de separação?

c. Em um conjunto de dados com separação quase linear, qual kernel tenderia a ser mais adequado? Justifique.


R:
a. O kernel utilizado no modelo SVM vai depender do tipo de padrão de dados que estamos tentando capturar. O SVM busca traçar um hiperplano (podendo ser uma linha, um plano ou um hiperplano propriamente dito) que melhor faça a divisão dos dados, o que é simples de imaginarmos se os dados possuem padrões lineares ou até mesmo exponenciais, porém fica bem mais difícil de abstrair se os dados não apresentam um padrão identificável. É daí que vem o kernel trick: é uma forma de projetar os dados para um espaço de maiores dimensões de forma com que a separação seja possível, porém sem a necessidade de explicitamente construir esse espaço. Ao mudarmos o kernel utilizado (linear, polinomial ou RBF) estamos mudando o tipo de hiperplano utilizado para a divisão de dados.

---

b. O C é o hiperparâmetro responsável pela flexibilização das margens, ou, em outras palavras, o peso que damos aos erros. Um C mais alto corresponde a margens mais estreitas e uma menor tolerência a erros, o que implica em maiores chances de overfitting e menor capacidade de generalização para novos dados. De forma contrária, um C pequeno corresponde a margens maiores e a uma maior tolerância a erros. Dessa forma, há um menor risco de overfitting e maior capacidade de generalização.

Já o gamma é o hiperparâmetro responsável por controlar a influência de cada ponto de dados nas fronteiras de decisão. Quando é um valor alto, a influência é muito restrita, o que faz com que as curvas da fronteira de decisão fiquem muito apertadas, aumentando o risco de overfitting. Em contrapartida, valores baixos fazem com que os pontos influenciem áreas muito grandes, o que se traduz em curvas mais suaves e uma melhor capacidade de generalização.

Modelos com gamma e C altos tendem a ser mais complexos, porém também com altas chances de overfitting e com uma capacidade bem baixa de generalização.

---

c. Em um conjunto de dados quase lineares, acredito que um kernel linear consiga capturar quase que a totalidade dos dados com um grau alto de acurácia. É a minha primeira escolha pois é um modelo mais simples de ser treinado, com menos hiperparâmetros a serem ajustados e que consegue capturar relações lineares (e quase lineares) com consistência e com baixo risco de overfitting.

## Questão 3 - Bagging vs Boosting:


a. Conceitue os métodos de ensemble Bagging e Boosting.

b. Dê um exemplo clássico de algoritmo de cada tipo.

c. Cite uma vantagem e uma desvantagem de cada abordagem no contexto de classificação.



R:

a. Bagging é um método de Machine Learning em que realizamos a amostragem aleatória de instâncias que farão parte do conjunto de treino, com a ressalva de que estes dados podem ser repetidos (inseridos no subconjunto mais de uma vez). A partir de cada subconjunto escolhido, um modelo diferente de ML é treinado, de forma paralela (o número de modelos treinados é determinado por um hiperparâmetro). O resultado final é decidido pela maior votação (no caso de problemas de classificação) ou na média dos valores encontrados entre os modelos independentes, no caso de problemas de regressão.

Já o Boosting é uma forma de treinar modelos em sequência, em que um modelo aprende com os erros do seu antecessor e vai se aprimorando iterativamente. A partir de um modelo inicial ruim/fraco, partimos para um segundo modelo, que irá trabalhar com os resíduos (erros) do modelo anterior, e assim por diante até chegarmos a um ponto de parada especificado. Dessa forma, o treinamento de modelos é realizado de forma consecutiva e sequencial.

---

b.

*   Bagging: Random Forest
*   Boosting: AdaBoost

---

c.

* Vantagens:
  - Bagging: Resistente a overfitting
  - Boosting: Reduz viés

* Desvantagens:
  - Bagging: Diminui a interpretabilidade do modelo
  - Boosting: Propenso a overfitting

## Questão 4 - SHAP e LIME:


a. O que são SHAP e LIME? Qual é o objetivo desses métodos no contexto de modelos preditivos?

b. Compare os dois métodos em termos de:

- Tipo de explicação gerada (global vs. local)

- Estabilidade das explicações

c. Cite uma situação em que o uso de SHAP seria preferível ao de LIME e justifique.


R:
a. SHAP (SHapley Addictive exPlanations) e LIME (Local Interpretable Model-agnostic Explanations) são ferramentas de interpretação de modelos de ML. Elas visam explicar como o modelo tomou as decisões e chegou a uma conclusão (predição final) e o que influenciou estas conclusões.

---

b. O LIME é a ferramenta de escolha quando queremos avaliar individualmente predições, pois só oferece interpretabilidade local (devido ao seu funcionamento). Porém apresenta outras limitações: não é um modelo estável, visto que a aleatoriedade na criação de perturbações durante o treinamento do modelo pode incorrer em explicações diferentes para instâncias parecidas e tem um desempenho melhor em modelos de complexidade menor.

Já o SHAP é uma ferramenta mais consistente e estável, pois seus valores de Shapley seguem os mesmos princípios da Teoria dos Jogos para todos os modelos. Também é uma ferramenta mais versátil, pois é compatível com vários tipos de modelos, inclusive os mais complexos como modelos de Deep Learning, e oferece interpretabilidade tanto local quanto global. Dessa forma, costuma ser uma ferramenta mais utilizada que o LIME.

---
c. Como mencionado anteriormente, o SHAP é melhor indicado para modelos mais complexos, onde precisamos ter um entendimento global e também local do funcionamento do modelo. Um exemplo seria em redes neurais usadas para reconhecimento facial, ou em situações de predições para diagnósticos médicos (onde necessitamos de modelos extremamente precisos).

## Questão 6 - Decision Tree e Random Forest


a. Treine os modelos de Árvore de Decisão e Random Forest.

b. Avalie os resultados com validação cruzada (5 folds) e F1-score.

c. Apresente os resultados e compare os modelos.


```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import f1_score, confusion_matrix

#lista para armazenar os resultados
results = []

#montando os modelos
models = {
    'Decision Tree': DecisionTreeClassifier(
        max_depth=30,
        min_samples_split=10,
        min_samples_leaf=10,
        random_state=37,
        ccp_alpha=0.01
    ),
    'Random Forest': RandomForestClassifier(
        n_estimators=100,
        max_depth=10,
        random_state=42,
        n_jobs=-1,
    )
}

plt.figure(figsize=(15, 6))
plot_idx = 1

for name, model in models.items():
    #Validação cruzada
    cv_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='f1', n_jobs=-1)
    f1_folds = [round(score, 3) for score in cv_scores]
    f1_mean = f"{np.mean(cv_scores):.3f} ± {np.std(cv_scores):.3f}"

    #fazendo as predições no conjunto de teste
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    f1_test = round(f1_score(y_test, y_pred), 3)

    #armazenando resultados
    results.append([
        name,
        f1_mean,
        f1_folds,
        f1_test
    ])

    #plotando a matriz de confusão
    cm = confusion_matrix(y_test, y_pred)
    plt.subplot(1, len(models), plot_idx)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Greens',
                xticklabels=['Negativo Predito', 'Positivo Predito'],
                yticklabels=['Negativo Real', 'Positivo Real'])
    plt.title(f'{name} Matriz de Confusão')
    plot_idx += 1

results_df = pd.DataFrame(
    results,
    columns=["Modelo", "F1 Médio (CV) ± Desvio Padrão", "F1 por Fold (CV)", "F1 Teste"]
)

#mostrando resultados
display(results_df)
plt.tight_layout()
plt.show()
```

R: Como era de se esperar, o modelo Random Forest teve um desempenho melhor do que a Decision Tree (ainda que a diferença seja pequena). Entretanto, ambos os modelos apresentam sinais de overfitting - há uma queda de 30% no F1-Score quando comparamos os resultados de treino e de teste. Porém, de uma forma geral, os desempenhos dos modelos foram semelhantes neste conjunto de dados. Talvez com um ajuste de hiperparâmetros conseguiríamos resultados melhores.

## Questão 8 - Ajuste de hiperparâmetros no SVM


a. Explore como os parâmetros C e gamma influenciam o desempenho do SVM.

b. Teste diferentes combinações e visualize as fronteiras de decisão em pelo menos dois cenários contrastantes.

c. Comente os efeitos observados na complexidade e no desempenho do modelo.


```python
from sklearn.model_selection import GridSearchCV

#grid dos parâmetros a serem testados
param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': [0.01, 0.1, 1, 10]
}

#grid search para iterar sobre os modelos
svm = SVC(kernel='rbf', random_state=42)
grid_search = GridSearchCV(svm, param_grid, cv=5, scoring='f1', n_jobs=-1, verbose=0)
grid_search.fit(X_train, y_train)

#guardando os resultados
svm_results = []

for params in grid_search.cv_results_['params']:
    model = SVC(kernel='rbf', **params, random_state=42)
    cv_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='f1', verbose=0)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    svm_results.append({
        'C': params['C'],
        'gamma': params['gamma'],
        'F1_CV_mean': np.mean(cv_scores),
        'F1_CV_std': np.std(cv_scores),
        'F1_test': f1_score(y_test, y_pred),
    })

#criando dataframe com os resultados
svm_results_df = pd.DataFrame(svm_results)

#dataframe final comparando todos os modelos do gridsearch
final_df = pd.DataFrame({
    'C': svm_results_df['C'],
    'gamma': svm_results_df['gamma'],
    'F1 CV (Média ± Desvio Padrão)': [f"{m:.3f} ± {s:.3f}" for m,s in zip(svm_results_df['F1_CV_mean'], svm_results_df['F1_CV_std'])],
    'F1 (Teste)': svm_results_df['F1_test'].round(3),
}).sort_values('F1 (Teste)', ascending=False)

# Return just the DataFrame
final_df
```

```python
from sklearn.decomposition import PCA

#pca para posterior visualização das fronteiras de decisão
pca = PCA(n_components=2)
X_train_2d = pca.fit_transform(X_train)
X_test_2d = pca.transform(X_test)
```

```python
from mlxtend.plotting import plot_decision_regions

#pegando o melhor e pior modelo com base no desempenho no grupo de teste
best_idx = svm_results_df['F1_test'].idxmax()
worst_idx = svm_results_df['F1_test'].idxmin()

#plotando as fronteiras de decisão
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
fig.suptitle('Fronteiras de Decisão dos Modelos SVM RBF')

#melhor modelo
best_model = SVC(kernel='rbf', **grid_search.best_params_, random_state=42)
best_model.fit(X_train_2d, y_train)
plot_decision_regions(X_train_2d, y_train.values, clf=best_model, ax=ax1)
ax1.set_title(f"Melhor Modelo (C={grid_search.best_params_['C']}, γ={grid_search.best_params_['gamma']})\n F1 (teste): {svm_results_df.loc[best_idx, 'F1_test']:.3f}")
ax1.legend(loc='lower right')

#pior modelo
worst_params = svm_results_df.loc[worst_idx, ['C', 'gamma']].to_dict()
worst_model = SVC(kernel='rbf', **worst_params, random_state=42)
worst_model.fit(X_train_2d, y_train)
plot_decision_regions(X_train_2d, y_train.values, clf=worst_model, ax=ax2)
ax2.set_title(f"Pior Modelo (C={worst_params['C']}, γ={worst_params['gamma']})\nF1 (teste): {svm_results_df.loc[worst_idx, 'F1_test']:.3f}")
ax2.legend(loc='lower right')

plt.tight_layout()
plt.show()
```

R: Podemos observar que valores gamma mais altos fizeram o modelo overfitar de forma absurda, o que corrobora sua definição teórica de que restringe a influência dos pontos de dados a um menor raio, fazendo curvas mais extremas. Dessa forma, temos um modelo mais complexo e com menor capacidade de generalização. Porém, nesse caso em específico, valores de C mais altos beneficiaram o modelo, o que faz sentido, pois os pontos de dados aparentam estar muito sobrepostos entre si, então margens mais estreitas ajudaram o modelo na melhor separação de classes.

```python
positive_indices = np.where(y_test == 1)[0] #apenas instâncias positivas

instance_idx = positive_indices[10] #número da instância que queremos averiguar dentro da classe positiva

test_shap_values = shap_explainer.shap_values(X_test)

#gráfico waterfall para a instância específica do conjunto de teste
shap.waterfall_plot(
    shap.Explanation(
        values=test_shap_values[instance_idx],
        base_values=shap_explainer.expected_value,
        data=X_test_df.iloc[instance_idx],
        feature_names=feature_names
    ),
    max_display=10,
    show=False
)

plt.title(f'Instância #{instance_idx} - Classe Positiva')
plt.tight_layout()
plt.show()
```

```python
#pegando os valores de SHAP da instância selecionada
base_log_odds = shap_explainer.expected_value
shap_sum_log_odds = test_shap_values[instance_idx].sum()
final_log_odds = base_log_odds + shap_sum_log_odds

#conversão de log odds para probabilidades
base_prob = 1 / (1 + np.exp(-base_log_odds))
final_prob = 1 / (1 + np.exp(-final_log_odds))

#convertendo os valores de SHAP para probabilidades
shap_effects_prob = []
for shap_val in test_shap_values[instance_idx]:
    effect = (1 / (1 + np.exp(-(base_log_odds + shap_val))) - base_prob)
    shap_effects_prob.append(effect)

#verificando probabilidades
print("Valores SHAP em probabilidades:")
print(f"Média das Predições (E[f(X)]): {base_prob:.3f}")
print(f"Predição Final (f(x)): {final_prob:.3f}")
print(f"Soma valores SHAP: {np.sum(shap_effects_prob)}")
```

R: Ao calcularmos a probabilidade da predição final do modelo, podemos ver que esta instância foi corretamente classificada como positiva, com uma certeza de praticamente 100%. Além disso, as duas principais features identificadas como mais importantes na análise geral também tiveram o maior impacto nesta instância, com a maior contribuição positiva vindo da feature classificada como mais importante (tanto pelo SHAP como pelo modelo - PageValues). Desta forma, a média das predições para a classe positiva foi de 0.561, o que nos mostra que o modelo está, em sua forma basal, tendendo à classe positiva e esta feature acrescentou, em muito, a probabilidade de classe positiva para esta instância. As outras features tiveram impactos variados, mas em escala bem menor e com tendência a se anularem entre si.

A explicação local da instância escolhida respeitou as conclusões observadas no gráfico SHAP global, evidenciando o impacto expressivo das features PageValues e Month nas probabilidades do modelo.

```python
import warnings
from lime.lime_tabular import LimeTabularExplainer

#Extraindo os pesos do LIME
def return_weights(exp, class_idx=1):
    """Extracts weights from LIME explanation for specified class"""
    exp_list = exp.as_map()[class_idx]
    exp_list = sorted(exp_list, key=lambda x: x[0])
    return [x[1] for x in exp_list]

#Calculando as importâncias das features das 1000 primeiras instâncias
weights = []
predict_fn = lambda x: model.predict_proba(pd.DataFrame(x, columns=X.columns))

for i in range(1000):
    x_instance = X_train_df.iloc[i]
    with warnings.catch_warnings():
        warnings.simplefilter("ignore")
        exp = lime_explainer.explain_instance(
            data_row=x_instance.values,
            predict_fn=predict_fn,
            num_features=len(X.columns)
        )
        weights.append(return_weights(exp, class_idx=1))

lime_weights = pd.DataFrame(data=weights, columns=X.columns)
abs_mean = lime_weights.abs().mean().sort_values(ascending=False)

plt.figure(figsize=(20, 10))

#Gráfico LIME
plt.subplot(1, 2, 1)
abs_mean.plot(kind='barh')
plt.xlabel('Mean |Weight|')
plt.title('Importância Absoluta Média (LIME)')
plt.gca().invert_yaxis()
plt.show()

#Gráfico SHAP
plt.subplot(1, 2, 2)
shap.summary_plot(shap_values, X_train, feature_names=feature_names, plot_type='bar', show=False)
plt.title('SHAP Feature Importance')
plt.xlabel('mean(|SHAP value|)')
plt.tight_layout()
plt.show()
```
