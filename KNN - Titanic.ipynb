## Exercício 1: A importância da generalização

Explique a diferença entre generalização e ajuste ao treinamento. Como saber se um modelo está generalizando bem? O que pode indicar que o modelo está apenas "memorizando" os dados de treinamento? Justifique sua resposta.

A generalização refere-se à capacidade do modelo de aplicar o que ele aprendeu durante o processo de treinamento para dados reais, ou seja, dados que não foram vistos durante o treinamento.
Já o ajuste nada mais é do que a avaliação do padrão identificado pelo modelo. Seus extremos chamam-se overfitting e underfitting.
Um modelo com boa capacidade de generalização é capaz de realizar predições com uma alta precisão, enquanto modelos sem capacidade de generalização possuem péssimo desempenho. Podemos utilizar métricas de avaliação (acurácia, f1, precisão, recall, etc) como formas de avaliação dos modelos.
Podemos suspeitar que o modelo está apenas memorizando os dados de treinamento quando obtemos um desempenho muito bom durante a fase de treinamento porém muito ruim nos dados de teste, o que nos leva a um modelo com baixa capacidade de generalização. Isso também pode ocorrer em modelos mais complexos, pois há maiores chances de "ajustarmos demais" os modelos (overfitting).

## Exercício 2: Overfitting e Underfitting

Defina overfitting e underfitting. Dê exemplos de situações em que ambos podem ocorrer ao treinar um modelo no dataset Titanic. Como você pode identificar se o seu modelo está sofrendo algum desses problemas?

- Overfitting: O modelo está ajustado demais para os dados e acaba por "decorá-los". Isso leva a um ótimo desempenho durante a fase de treinamento porém a um péssimo desempenho na fase de teste.
- Underfitting: O modelo está simples demais para o problema estudado. Seu desempenho é ruim em todas as etapas do processo, pois não consegue detectar padrões nos dados.

No dataset Titanic, se utilizarmos o algoritmo KNN para predizer se um passageiro sobreviveu ou não, pode acontecer de utilizarmos um K muito alto, o que aumenta a capacidade de generalização do modelo, porém gera um underfitting. De forma inversa, se utilizarmos um k muito baixo, estamos adaptando demais o modelo aos dados de treinamento, gerando um overfitting.

Para identificar se o modelo está com um desempenho satisfatório, podemos lançar mão das mais diversas métricas de avaliação, como por exemplo acurácia, precisão, recall, f1, etc.

##Exercício 3: Boas Práticas de Pré-processamento de Dados

Liste e explique que boas práticas de pré-processamento de dados devem ser realizadas antes de treinar um modelo de classificação. Descreva as principais ideias por trás de cada uma destas práticas.

Etapas do pré-processamento de dados:
1. Limpeza de dados: Devemos sempre checar por dados faltantes, incompletos, inconsistentes ou irrelevantes ao estudo e tratá-los de forma adequada. Apagar entradas faltantes ou irrelevantes e preencher dados incompletos é essencial para se melhorar o desempenho do modelo. Os dados faltantes (NaN) não são reconhecidos pelos modelos, então não conseguimos realizar treinamentos com esses dados. Apagar entradas irrelevantes ajuda o modelo a ser mais preciso, pois elimina potenciais fatores de confusão.
2. Conversão de dados categóricos para numéricos: Frequentemente encontramos features que pertencem à uma classe diferente da numérica. É essencial que convertamos esses dados para numéricos, visto que o modelo só sabe trabalhar com números, e não textos (strings).
3. Padronização dos dados: Mesmo depois que todos os dados pertençam à categoria numérica, pode acontecer de trabalharmos com ordens de grandeza extremamente diferentes, porém o modelo não sabe disso. Ele irá tratar números maiores como sendo "mais importantes", atribuindo um maior peso a esses dados quando comparados a dados menores. Entretanto, não podemos dizer que "18 anos" é menos importante do que "R$3452,56". Dessa forma, a padronização força os dados a ficarem na mesma ordem de grandeza (escala), o que elimina esse viés de seleção na etapa de treinamento.

##Exercício 5: Preparação dos Dados

Escreva um código Python para limpar e preparar os dados do Titanic para a criação de um modelo de classificação binária (sobreviver ou não). Considere a imputação de dados faltantes, conversão de variáveis categóricas e normalização (se necessário).

```python
import pandas as pd

#importando o dataset
df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv', index_col="PassengerId")

#leitura inicial dos dados
df
```

```python
#checando as entradas sem valores
nan_check = df.isna().sum()
nan_check
```

```python
#checando a divisão dos sobreviventes x não sobreviventes no dataset
survival_check = df.groupby('Survived').size()
survival_check
```

```python
import matplotlib.pyplot as plt
plt.pie(survival_check, labels=['Não Sobreviventes', 'Sobreviventes'], autopct='%1.1f%%')
plt.show()
```

```python
#muitos dados faltantes em cabines (687 de 891 entradas), melhor descartarmos esta coluna
treated_df = df.drop('Cabin', axis=1)
treated_df
```

```python
#vamos completar as idades faltantes com a mediana das idades dos passageiros (melhor do que média pois haverá apenas valores inteiros)
treated_df['Age'] = treated_df['Age'].fillna(treated_df['Age'].median())
treated_df
```

```python
#O nome do passageiro e o número do ticket também não parecem ser relevantes ao estudo, então vamos retirá-las do df
treated_df = treated_df.drop('Name', axis=1)
treated_df = treated_df.drop('Ticket', axis=1)
treated_df
```

```python
#agora, vamos transformar as variáveis categóricas em numéricas. precisamos fazer isto pras colunas 'sex' e 'embarked'
#gêneros: female = 0 e male = 1 // portos: C = 0, Q = 1, S = 2

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
treated_df['Sex'] = le.fit_transform(treated_df['Sex'])
treated_df['Embarked'] = le.fit_transform(treated_df['Embarked'])
treated_df
```

```python
#finalmente podemos separar nossas features e nosso target

X = treated_df.drop('Survived', axis=1)
y = treated_df['Survived']

#checando nossas features
X
```

```python
#embora os dados estejam praticamente na mesma escala, vamos escaloná-los para ter certeza de que não haverá desbalanço entre as features
#optei pelo standardScaler por não achar nada de especial sobre os dados, então nenhum cuidado adicional parece ser necessário

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

##Exercício 6: Construção do Modelo com KNN

Construa um modelo de K-Nearest Neighbors (KNN) utilizando o dataset Titanic para prever se o passageiro sobreviveu ou não. Utilize o Scikit-Learn e divida os dados em treino e teste.

```python
#finalmente podemos separar os conjuntos de treino e teste para montarmos nosso modelo

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
#agora vamos aplicar o algoritmo KNN. mantive o k padrão sugerido (5) para a primeira tentativa

from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
y_pred
```

```python
#por fim, vamos testar a acurácia do modelo através da comparação entre o y predito e o y real

from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print(f"Acurácia de {(accuracy * 100):.2f}%")
```

## Exercício 7: Ajustando o Modelo

Após construir o modelo inicial com KNN, ajuste o valor de k. Teste diferentes valores e determine qual resulta no melhor desempenho.

```python
#será que conseguimos a melhor acurácia possível? vamos testar outros valores para o k e checar as diferentes acurácias

for k in range(1, 50, 2):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Acurácia de {(accuracy * 100):.2f}% para k = {k}")
```

Em conclusão: 5 não parece ser o melhor valor para k neste modelo. Se usarmos k = 15 ou k = 29, conseguimos uma acurácia de 81,56%. Dentre os dois, eu optaria pelo k = 15 para reduzir o risco de underfitting.

## Exercício 8: Avaliação do Modelo com Acurácia

Utilizando o modelo KNN, calcule a acurácia do seu modelo no conjunto de teste. Explique se a acurácia é uma boa métrica para avaliar o desempenho do modelo no caso do Titanic. Quais são as limitações da acurácia para datasets desbalanceados?

```python
#vamos mudar nosso k para 15

knn = KNeighborsClassifier(n_neighbors=15)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Acurácia de {(accuracy * 100):.2f}%")
```

No caso do titanic, a acurácia pode ser uma métrica adequada para medir o desempenho do modelo pois os dados estão balanceados (549 não sobreviventes (aprox 62%) para 342 sobreviventes(aprox 38%)), então é praticamente garantido que cada uma das classes apareça tanto no conjunto de teste quanto no conjunto de treinamento.

Entretanto, no caso de datasets desbalanceados, a acurácia isolada é uma métrica falha, pois não há como garantirmos que todas as classes estejam presentes nos dois grupos. Dessa forma, o modelo pode ter uma acurácia falseada, pois pode ser treinado para predizer que todas as ocorrências de uma instância são verdadeiras - e acertar - caso elas sejam responsáveis por 90% dos dados.
A validação cruzada é uma boa forma de controlar o problema, assim como a utilização de outras métricas em conjunto com a acurácia.

##Exercício 9: Métricas de Classificação: Precisão, Recall e F1-Score

Calcule as métricas de precisão, recall e F1-score para o modelo KNN. Explique por que essas métricas podem ser mais informativas do que a acurácia, especialmente para um dataset desbalanceado como o Titanic.

```python
from sklearn.metrics import precision_score, recall_score, f1_score

precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Precisão: {(precision*100):.2f}")
print(f"Recall: {(recall*100):.2f}")
print(f"F1-Score: {(f1*100):.2f}")
```

Embora eu não considere o titanic um dataset desbalanceado, a precisão, o recall e o F1-Score são métricas mais informativas do que a acurácia isolada pois a precisão mede o quanto dos verdadeiros preditos pelo modelo são realmente verdadeiros, o recall mede a sensibilidade do modelo (quantas instâncias verdadeiras reais foram identificadas) e o F1-Score é a média harmônica entre as duas métricas anteriores. Dessa forma, conseguimos avaliar o modelo de uma forma mais completa.

## Exercício 10: Analisando o Impacto de Outliers

Verifique a presença de outliers nas variáveis numéricas do dataset Titanic. Crie um gráfico para mostrar a distribuição de variáveis como Idade e Fare e discuta como a presença de outliers pode afetar o desempenho do modelo.

```python
import seaborn as sns
import matplotlib.pyplot as plt

#descobrindo as idades máxima e mínima
min_age = treated_df['Age'].min()
max_age = treated_df['Age'].max()

#descobrindo o intervalo interquantil
Q1 = treated_df['Age'].quantile(0.25)
Q3 = treated_df['Age'].quantile(0.75)

sns.boxplot(x=treated_df['Age'])
plt.title('Distribuição de Idade')

# Adicionando os valores ao gráfico de idade
plt.text(min_age, -0.3, f'{min_age:.2f}', ha='center', va='top', color='red')
plt.text(max_age, -0.3, f'{max_age:.2f}', ha='center', va='top', color='red')
plt.text(Q1, -0.3, f'{Q1}', ha='center', va='top', color='red')
plt.text(Q3, -0.3, f'{Q3}', ha='center', va='top', color='red')
plt.show()

#repetindo o processo para o outro gráfico
min_fare = treated_df['Fare'].min()
max_fare = treated_df['Fare'].max()

Q1 = treated_df['Fare'].quantile(0.25)
Q3 = treated_df['Fare'].quantile(0.75)

sns.boxplot(x=treated_df['Fare'])
plt.title('Distribuição da preço da entrada')

plt.text(min_fare, -0.3, f'{min_fare:.2f}', ha='center', va='top', color='red')
plt.text(max_fare, -0.3, f'{max_fare:.2f}', ha='center', va='top', color='red')
plt.text(Q1, -0.20, f'{Q1:.2f}', ha='right', va='bottom', color='red')
plt.text(Q3, -0.20, f'{Q3:.2f}', ha='left', va='bottom', color='red')
plt.show()
```

Através dos gráficos podemos verificar que as duas categorias (Idade e Preço) possuem vários outliers. Os outliers são dados que diferem muito da normalidade, ou seja, se diferenciam drasticamente das outras instâncias. Como exemplos temos o passageiro com 80 anos ou a pessoa que pagou 512.33UI na entrada do navio. Esses valores são importantes pois têm o potencial causar anomalias no comportamento de predição dos modelos, visto que alteram (e muito) a distribuição, média e mediana dos dados. Dessa forma, podem criar vieses durante o treinamento dos modelos. Normalmente, declara-se como um outlier qualquer valor 1,5xIQR menor que o primeiro quartil ou 1,5xIQR maior que o terceiro quartil.

## Exercício 11: Regressão no Titanic

Poderíamos usar um modelo de regressão para prever a probabilidade de um passageiro sobreviver ao invés de uma classificação binária (sobreviveu ou não)? Qual seria a diferença em termos de avaliação do modelo? Compare a avaliação de um modelo de regressão com um modelo de classificação no contexto do Titanic.

Sim. O modelo de regressão seria responsável por prever a probabilidade de um dado passageiro sobreviver. A resposta seria dada na forma de um número contínuo de 0 a 1, com 0 sendo 0% de chances e 1 sendo 100% de chances de sobrevivência.
Para avaliação do modelo, teríamos que lançar mão de métricas de avaliação como erro médio absoluto, erro médio quadrático, raiz do erro quadrático médio, etc.

```python
#vamos treinar um modelo de regressão usando o df já tratado
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
y_pred
#podemos ver que os três primeiros passageiros provavelmente não sobreviveram
```

```python
#vamos testar as métricas

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"MAE: {mae:.2f}")
print(f"MSE: {mse:.2f}")
print(f"R2: {r2:.2f}")
```

```python
#colocando em gráfico

#definindo o threshold
threshold = 0.5

# Cria o gráfico de dispersão
plt.scatter(y_test, y_pred, alpha=0.5)

# Adiciona a linha horizontal representando o threshold
plt.axhline(y=threshold, color='red', linestyle='-')

# Configura os rótulos dos eixos e o título
plt.xlabel("Valor Real (0: Não sobreviveu, 1: Sobreviveu)")
plt.ylabel("Probabilidade Prevista")
plt.title("Comparação do Modelo de Regressão com os Dados Reais")

# Exibe o gráfico
plt.show()
```
