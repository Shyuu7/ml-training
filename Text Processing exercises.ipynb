# Parte 2 Seleção de Features

## Questão 5:

Considere o Breast Cancer Dataset. Proponha uma estratégia de filtragem para selecionar um conjunto de features mais significativas e justifique sua resposta.

R: As features que devemos considerar devem ser as estatisticamente relevantes ao resultado, ou seja, as que têm maior chance de interferirem diretamente no diagnóstico (seja ele  positivo ou negativo). O cálculo do p é uma forma matemática de se mensurar o tamanho do impacto do dado no desfecho estudado. Por via de regra, p < 0.05 significa que o dado é estatisticamente relevante ao estudo, enquanto valores acima disso significam que não foi encontrada uma relação direta entre o dado e o desfecho.

```python
#importando o dataset
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
X = data.data
y = data.target
feature_names = data.feature_names

data_df = pd.DataFrame(X, columns=feature_names)
data_df.head()
```

```python
#aplicando o teste estatístico para descobrir o p de cada feature
from sklearn.feature_selection import f_classif
f_statistic, p_values = f_classif(X, y)


#listando o p por ordem de grandeza
feature_importance = pd.DataFrame({'Feature': feature_names, 'p-value': p_values})
feature_importance = feature_importance.sort_values(by=['p-value'])
feature_importance
```

```python
#listando as features selecionadas
selected_features = feature_importance[feature_importance['p-value'] < 0.05]['Feature'].values
selected_features
```

Assim, 25 das 30 colunas são consideradas estatisticamente relevantes. Desconsideramos "fractal dimension error", "smoothness error", "mean fractal dimension", "texture error" e "symmetry error".

## Questão 6:

Repita a Questão 5, mas agora utilizando o método Wrapper para selecionar as features. Compare os resultados obtidos com os da abordagem de filtragem.

```python
from sklearn.feature_selection import RFECV
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold

estimator = LogisticRegression(max_iter=10000)

cv = StratifiedKFold(5)
rfecv = RFECV(estimator, step=1, cv=cv, scoring='accuracy')

rfecv.fit(X, y)

# Pegando as features selecionadas
selected_features_rfecv = [feature_names[i] for i in range(len(feature_names)) if rfecv.support_[i]]

# Mostrando o resultado
print("Número de features: %d" % rfecv.n_features_)
print("Features selecionadas (RFECV):", selected_features_rfecv)
```

Nesse caso, podemos ver que o wrapper otimizou ainda mais o número de features para seleção quando comparado à filtragem (25 > 20). Ao juntar regressão logística com a validação cruzada, estamos testando e ranqueando todas as features para selecionar as melhores. Desta forma, ao fazer um estudo mais detalhado dos dados, é esperado que seu desempenho seja melhor do que o filtro adotado na questão anterior, embora computacionalmente mais caro, pois possue mais etapas e mais testes realizados.

## Questão 7:

Descreva o funcionamento do método de seleção de features conhecido como Embedding. Explique por que ele é computacionalmente mais barato que o Wrapper, justificando sua resposta.

O Embedding é o método de seleção de features que acontece inerentemente ao processo de treinamento do modelo de aprendizagem de máquina. Ao receber dados de entrada, os modelos irão transformá-los em representações vetoriais das features originais, mas com dimensionalidades reduzidas, assim dando maior peso às features mais relevantes.

É computacionalmente mais barato que o Wrapper pois não são necessárias etapas adicionais ao processo de treinamento; ele ocorre de forma inerente. Também ajuda o fato de ocorrer a redução da dimensionalidade dos dados, o que auxilia o modelo a buscar mais rapidamente o subconjunto de features otimizado.

## Questão 8:

Utilize o método Bag-of-Words para transformar o texto acima em um vetor. Apresente a vetorização obtida.

```python
from sklearn.feature_extraction.text import CountVectorizer
text = ['A alma é, pois, imortal; renasceu repetidas vezes na existência e contemplou todas as coisas existentes e por isso não há nada que ela não conheça! Não é de espantar que ela seja capaz de evocar à memória a lembrança de objetos que viu anteriormente, e que se relacionam tanto com a virtude como com as outras coisas existentes. Toda a natureza, com efeito, é uma só, é um todo orgânico, e o espírito já viu todas as coisas; logo, nada impede que ao nos lembrarmos de uma coisa – o que nós, homens, chamamos de “saber” – todas as outras coisas acorram imediata e maquinalmente à nossa consciência.']
vect = CountVectorizer()
vect.fit(text)
print(vect.vocabulary_)

bag_of_words = vect.transform(text)
print("Vetor:\n{}".format(bag_of_words.toarray()))
```

## Questão 9:

Utilize o método Bag-of-n-Grams para vetorizar o texto e apresente:

- O número de unigrams.
- O número de bigrams.
- O número de trigrams.


```python
cv1 = CountVectorizer(ngram_range=(1, 1)).fit(text)
print("Unigrams: {}".format(len(cv1.vocabulary_)))

cv2 = CountVectorizer(ngram_range=(2, 2)).fit(text)
print("Bigrams: {}".format(len(cv2.vocabulary_)))

cv3 = CountVectorizer(ngram_range=(3, 3)).fit(text)
print("Trigrams: {}".format(len(cv3.vocabulary_)))
```

##Questão 10:

Utilizando o pacote NLTK, remova as stopwords do texto e apresente o resultado obtido.

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download("stopwords")
nltk.download("punkt_tab")
```

```python
words = word_tokenize(str(text))

stop_words = set(stopwords.words("portuguese"))
text_no_stopwords = [word for word in words if word.lower() not in stop_words]
str(text_no_stopwords)
```

## Questão 11:

Utilize o NLTK para aplicar stemming no texto. Apresente o resultado obtido.

```python
from nltk.stem import RSLPStemmer
nltk.download("rslp")
import string

stemmer = RSLPStemmer()
text_no_punctuation = str(text_no_stopwords).translate(str.maketrans("", "", string.punctuation))
word_tokens = word_tokenize(text_no_punctuation)

print("{0:20}{1:20}".format("--Palavra--","--Stem--"))
for word in word_tokens:
    print ("{0:20}{1:20}".format(word, stemmer.stem(word)))
```
